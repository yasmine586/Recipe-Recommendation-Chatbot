```python
import spacy
import re
import pandas as pd

nlp = spacy.load("en_core_web_sm")
```

```python
def extract_ingredients(ingredient_text):
    """
    Extract clean ingredient names ready for embedding.
    Returns a plain string of space-separated ingredients.
    """
    if pd.isna(ingredient_text):
        return ""
    
    # Common measurement units to skip
    units = {'c', 'c.', 'cup', 'cups', 'tbsp', 'tsp', 'oz', 'oz.', 'lb', 'lb.', 
             'pkg', 'pkg.', 'jar', 'can', 'package', 'g', 'ml', 'l', 'qt', 'pt',
             'box', 'container', 'carton', 'slices', 'slice','small','large','medium'}
    
    # Parse the string and split ingredients
    ingredient_text = ingredient_text.strip('[]"')
    ingredients = re.split(r'",\s*"', ingredient_text)
    
    result = []
    for ing in ingredients:
        ing = ing.strip('"')
        doc = nlp(ing)
        
        # Extract nouns, skip measurement units
        nouns = [token.text.lower() for token in doc 
                 if token.pos_ in ['NOUN', 'PROPN','ADJ'] and token.text.lower() not in units]
        
        if nouns:
            # Join nouns and clean up punctuation
            clean_ingredient = ' '.join(nouns)
            # Remove periods, commas, semicolons
            clean_ingredient = re.sub(r'[.,;]+', '', clean_ingredient)
            clean_ingredient = clean_ingredient.strip()
            
            if clean_ingredient:
                result.append(clean_ingredient)
    
    # Return as space-separated STRING (not list!)
    return ', '.join(result)
```

```python
df=pd.read_csv("recipes_data.csv",nrows=100)
df['ingredients_flat'] = df['ingredients'].apply(extract_ingredients)

```
